# Taking only genres names as required

```python
def convert_genres(genres):
    if isinstance(genres, str):
        return eval(genres)
    return genres

def extract_genre_names(genres_list):
    return [genre['name'] for genre in genres_list] if isinstance(genres_list, list) else []


df['genres'] = df['genres'].apply(convert_genres)


df['genres'] = df['genres'].apply(extract_genre_names)


df['genres_name'] = df['genres'].apply(', '.join)
```

# Taking only keywords name as required

```python
def convert_keywords(keywords):
    if isinstance(keywords, str):
        return eval(keywords)
    return keywords

def extract_keyword_names(keywords_list):
    return [keyword['name'] for keyword in keywords_list] if isinstance(keywords_list, list) else []

df['keywords'] = df['keywords'].apply(convert_keywords)

df['keywords_name'] = df['keywords'].apply(extract_keyword_names)

df['keywords_name'] = df['keywords_name'].apply(', '.join)
```

# Taking only production_companies name as required

```python
def convert_production_companies(companies):
    if isinstance(companies, str):
        return eval(companies)
    return companies

def extract_production_company_names(companies_list):
    return [company['name'] for company in companies_list] if isinstance(companies_list, list) else []

df['production_companies'] = df['production_companies'].apply(convert_production_companies)

df['production_companies_name'] = df['production_companies'].apply(extract_production_company_names)

df['production_companies_name'] = df['production_companies_name'].apply(', '.join)
```

# Taking only spoken_languages name as required

```python
def convert_spoken_languages(languages):
    if isinstance(languages, str):
        return eval(languages)
    return languages

def extract_first_spoken_language(languages_list):
    if isinstance(languages_list, list) and len(languages_list) > 0:
        return languages_list[0]['name']
    return ''

df['spoken_languages'] = df['spoken_languages'].apply(convert_spoken_languages)

df['spoken_languages_name'] = df['spoken_languages'].apply(extract_first_spoken_language)
```

# Task -7: How are popularities of movies related with the movie budgets? Are they correlated or totally uncorrelated with each other? Write the interpretation of your analysis.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='popularity', y='budget', data=df)
plt.title('Popularity vs. Budget')
plt.xlabel('Popularity')
plt.ylabel('Budget')
plt.show()
```
